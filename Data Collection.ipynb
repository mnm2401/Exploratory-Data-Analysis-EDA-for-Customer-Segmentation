```python
import pandas as pd

import random
```

```python
customer_data = []
for i in range(500):
    customer_id = random.randint(1001, 6000)
    age = random.randint(20, 65)
    gender = random.choice(["Male", "Female"])
    marital_status = random.choice(["Married", "Single", "Divorced"])
    annual_income = random.randint(30000, 120000)
    total_purchases = random.randint(10, 50)
    preferred_category = random.choice(["Electronics", "Appliances", "Home Goods"])
    customer_data.append([customer_id, age, gender, marital_status, annual_income, total_purchases, preferred_category])

print(customer_data)
```

```python
dataset=pd.DataFrame(customer_data)
dataset
```

```python
dataset.columns=['customer_id','age', 'gender', 'marital_status', 'annual_income', 'total_purchases', 'preferred_category']
dataset
```

```python
dataset.to_csv('Customer_Dataset.csv')
```
